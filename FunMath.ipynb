```python
def extractdigits(num):
  if num >= 10:
    return (*extractdigits(num//10), num % 10)
  return (num,)

def combinedigits(num):
  if len(num) > 0:
    return (num.pop() + 10*combinedigits(num))
  return 0

def jankmath(ntuple):
  numtuple = extractdigits(ntuple[0])
  dentuple = extractdigits(ntuple[1])
  numlist = [i for i in numtuple if i not in dentuple]
  denlist = [i for i in dentuple if i not in numtuple]
  num = combinedigits(numlist)
  den = combinedigits(denlist)
  if den == 0:
    den += 0.00001
  return (num/den)
```

```python
def jankcheck(ntuple):
  numtuple = extractdigits(ntuple[0])
  dentuple = extractdigits(ntuple[1])
  numlist = [i for i in numtuple if i not in dentuple]
  denlist = [i for i in dentuple if i not in numtuple]
  return (len(numlist) != len(numtuple) or len(denlist) != len(dentuple)) and set(numtuple) != set(numlist + [0])
```

```python
jankcheck((95,19))
```

```python
import seaborn as sns
import numpy as np
```

```python
X,Y = np.mgrid[1:10000:1, 1:10000:1]
```

```python
flatgrid = zip(X.flatten(),Y.flatten())
xdata = [i[0]/i[1] for i in zip(X.flatten(),Y.flatten())]
ydata = [jankmath(i) for i in zip(X.flatten(),Y.flatten())]
endlist = [(i[0],i[2]) for i in zip(xdata, ydata, flatgrid) if i[0] == i[1] and jankcheck(i[2])]
```
